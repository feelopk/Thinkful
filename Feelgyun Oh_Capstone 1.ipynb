## Non-recyclable plastic import ban of China and its impacts

**INTRODUCTION**

China had declared that they won't import non-recyclable plastic waste from 2017. It is quite astonishing that China had been the most plastic waste import country since 1998.
We are able to find that there are four types of plastic wastes in the HS code. However, there are only classified by material, not by its recyclability. In this reason, here I am going to assume that all the trading waste plastics were considered as non-recyclable plastic.

Let's dive into waste plastic trade data issued by UN. 
First of all, it is better to find out how much plastics had been imported from 2011 to 2018 to China or Hongkong. Because China and Hongkong is a main actor of this analysis. Also it is better for you to know at the beginning of this journey that China has accustomed to import goods through Hongkong which is located seashore and transported imported things to inland by trains. For example, in 2018, China was not specified in the UN trade data but we already know that China uses Hongkong as their window of trades.

Now we all ready to embark on the journey! Let's move!

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

year = ['2011', '2012', '2013', '2014', '2015', '2016', '2017', '2018']
import_y = [33053684921, 32800594466, 30548476417, 32759977559, 30571976977, 30801914356, 27422402726, 8691213956]
CHN_y = [16768379852, 17755533284, 15762607164, 16508494578, 14708458120, 14694351058, 11657499768, 1196092726]
CHN_y_expected = [0, 0, 0, 0, 0, 0, 0, 2703473043]
ind = [x for x, _ in enumerate(year)]

fig = plt.figure(figsize=(15,5))

plt.subplot(1,2,1)
plt.bar(year, import_y, color='blue')
plt.xlabel('Import Year')
plt.ylabel('Amount (kg)')
plt.title('Plastic Waste Import Amount')

plt.subplot(1,2,2)
plt.bar(ind, CHN_y, color='red')
plt.bar(ind, CHN_y_expected, color='blue', bottom=CHN_y)
plt.xticks(ind, year)
plt.xlabel('Import Year')
plt.ylabel('Amount (kg)')
plt.title('Plastic Waste Import Amount of China')
plt.show()
```

Now, you can naturally find out that there was drastically decreasing in 2018. Surprisingly, that is not my mistake. 

Actually, in 2018, there was big decreasing on plastic waste trade all over the world. The left bar chart shows that overall plastic waste trade decreased drastically in 2018, almost a one-third amount in comparison to former years. However, the interesting thing is on the right side chart. This chart shows that the plastic waste import amount in China from 2011 to 2018. Especially, in 2018, China just comprised a one-tenth amount of that year's trade quantity. This is a very odd phenomenon because every year China had imported almost half the amount of whole plastic waste on average.

At this moment, let's set aside that why the waste plastic trade decreased in 2018. Briefly speaking, there were big movements on reducing plastic wastes all over the world. For instance, Starbucks Co. tried to use not plastic straws and made a transition from plastic things to paper things. Then, we can cast a doubt that where are the rest of wastes gone? (the blue one)

Now, it is time to check up the rest of the amount of plastic waste went out of China.

First of all, let's examine which countries did import most of the plastic wastes from all over the world. I am going to extract the countries which were highly ranked as waste plastic countries in 2017 and 2018.

Here is the 2017 data.

```python
df2017 = pd.read_csv('2017_Plastic Waste Data.csv')
df2017 = df2017.sort_values(by='Netweight', ascending=False)
df2017_CHN = df2017[df2017['Reporter ISO']=='CHN']
df2017_CHN_import = df2017_CHN[df2017_CHN['Trade Flow']=='Import']
df2017_CHN_import['Netweight'].sum()
df2017_HKG = df2017[df2017['Reporter ISO']=='HKG']
hk_list = df2017[df2017['Reporter ISO']=='HKG']
hk_list_index = hk_list.index.tolist()
df2017_noHKG = df2017.drop(hk_list_index)
df2017_import = df2017_noHKG[df2017_noHKG['Trade Flow']=='Import']
df2017_import['Weight Proportion'] = df2017_import['Netweight'] / df2017_import['Netweight'].sum()

count = 0
sum = 0

for w in df2017_import['Weight Proportion']:
    sum = sum + w
    count = count + 1
    if 0.7 < sum < 1:
        break
print(sum, count)

df2017_import[:12]
```

In 2017, the highly ranked countries are in sequence, China, Slovenia, Netherlands, Malaysia, Germany and USA.

Now, let's check the 2018 data.

```python
df2018 = pd.read_csv('2018_Plastic Waste Data_refined.csv')
df2018 = df2018.sort_values(by='Netweight', ascending=False)
df2018_import = df2018[df2018['Trade Flow']=='Import']
df2018_import['Weight Proportion'] = df2018_import['Netweight'] / df2018_import['Netweight'].sum()

count = 0
sum = 0

for w in df2018_import['Weight Proportion']:
    sum = w + sum
    count = count + 1
    if 0.5 < sum < 1:
        break
print(sum, count)

df2018_import[:13]
```

In 2018, the high ranked countries are China, Thailand, Indonesia, India, Belgium, the USA, and Germany.
It is quite interesting that the Top 7 countries taken up over 50% of plastic wastes import in the world. In addition, we can recognize that some countries appreared newly in 2018: Thailand, Indonesia, India and Belgium. Let's looking into later as for the newly emerging countries.

Here I give some neat pie chart to help you recognize how much the countries taking up the portion of the waste plastic import easily.

```python
other = df2017_import[13:]['Weight Proportion'].sum()
df2017_half = df2017_import[:13].groupby('Reporter').sum()
df2017_half.loc[6] = other
df2017_half_WP = df2017_half['Weight Proportion']
df2017_half_WP = df2017_half_WP.rename(index={6:'Others'})
label_2017 = np.array(df2017_half_WP.index)

other = df2018_import[13:]['Weight Proportion'].sum()
df2018_half = df2018_import[:13].groupby('Reporter').sum()
df2018_half.loc[7] = other
f2018_half_WP = df2018_half['Weight Proportion']
df2018_half_WP = df2018_half_WP.rename(index={7:'Others'})
label_2018 = np.array(df2018_half_WP.index)

plt.figure(figsize=(24,6))
plt.subplot(1,2,1)
plt.pie(df2017_half_WP, labels=label_2017, autopct='%1.1f%%')
plt.axis('equal')
plt.title('2017 Top Import Countries')
plt.legend(df2017_half_WP.index)

plt.subplot(1,2,2)
plt.pie(df2018_half_WP, labels=label_2018, autopct='%1.1f%%')
plt.axis('equal')
plt.title('2018 Top Import Countries')
plt.legend(df2018_half_WP.index, loc='lower left')
plt.show()
```

When it comes to the emerging import countries in 2018 are Thailand, Indonesia, India, and Belgium as I mentioned above. Before we declare that those countries are budding plastic waste countries, let's look up the former import amount data of those countries.

```python
df = pd.read_csv('total.csv')
df_import = df[df['Trade Flow']=='Import']
df_import = df_import.sort_values(by='Year', ascending=True)

# Thailand Case
df_Thailand = df_import[df_import['Reporter']=='Thailand'].groupby('Year').aggregate(np.sum)
np.mean(df_Thailand['Netweight'][:6])

# Indonesia Case
df_Indonesia = df_import[df_import['Reporter']=='Indonesia'].groupby('Year').aggregate(np.sum)
np.mean(df_Indonesia['Netweight'][:7])

# India Case
df_India = df_import[df_import['Reporter']=='India'].groupby('Year').aggregate(np.sum)
df_India['Netweight'][:7].mean()

# Belgium Case
df_Belgium = df_import[df_import['Reporter']=='Belgium'].groupby('Year').aggregate(np.sum)
df_Belgium['Netweight'][:7].mean()

# Plot the overall year import amount
df_Thailand_plot = df_Thailand['Netweight']
Thailand_mean = np.array([102572086, 102572086, 102572086, 102572086, 102572086, 102572086, 102572086])

df_Indonesia_plot = df_Indonesia['Netweight']
Indonesia_mean = np.array([225149006, 225149006, 225149006, 225149006, 225149006, 225149006, 225149006, 225149006])

df_India_plot = df_India['Netweight']
India_mean = np.array([381929237, 381929237, 381929237, 381929237, 381929237, 381929237, 381929237, 381929237])

df_Belgium_plot = df_Belgium['Netweight']
Belgium_mean = np.array([548629667, 548629667, 548629667, 548629667, 548629667, 548629667, 548629667, 548629667])

year_7 = np.array(df_Thailand_plot.index)
year_8 = np.array(df_Belgium_plot.index)

plt.figure(figsize=(18,10))

plt.subplot(2,2,1)
plt.plot(df_Thailand_plot, marker='o', color='b', linewidth=2)
plt.plot(year_7, Thailand_mean, linestyle='--', color='k')
plt.xlabel('Year')
plt.ylabel('Amount')
plt.title('Thailand Import Amount')

plt.subplot(2,2,2)
plt.plot(df_Indonesia_plot, marker='o', color='r', linewidth=2)
plt.plot(year_8, Indonesia_mean, linestyle='--', color='k')
plt.xlabel('Year')
plt.ylabel('Amount')
plt.title('Indonesia Import Amount')

plt.subplot(2,2,3)
plt.plot(df_India_plot, marker='o', color='g', linewidth=2)
plt.plot(year_8, India_mean, linestyle='--', color='k')
plt.xlabel('Year')
plt.ylabel('Amount')
plt.title('India Import Amount')

plt.subplot(2,2,4)
plt.plot(df_Belgium_plot, marker='o', color='purple', linewidth=2)
plt.plot(year_8, Belgium_mean, linestyle='--', color='k')
plt.xlabel('Year')
plt.ylabel('Amount')
plt.title('Belgium Import Amount')
plt.show()
```

Now, It is clear that the expected emerging countries are narrowed and just two countries are entitled to get the name of 'Budding Waste Plastic Import Country'. Those are Thailand and Indonesia

Thailand and Indonesia cases are showing a very similar pattern of the graph. Their average import amount which is drawn black dashed line clearly assure that they hadn't imported quite a big amount of waste plastic before 2018. Even Thailand didn't import waste plastic in 2017 at all. However, at this point, we could presume that after banning of China, they absorbed some portion of waste plastic which would have imported to China.

Let's look thoroughly about why India and Belgium are not suitable for getting entitled. 

First, India case, the import amount of waste plastic quite fluctuated from time to time. However, the average line(black line in the grid) shows that the variations are understandable. This case could account for the Belgium case also. 

In the Belgium case, the import amount of waste plastic was decreased in 2018, compared to the former year, even though the country was highly ranked. And from 2016, the import amount decreased. With this graph, Belgium has been trying to reduce to import plastic waste from 2016.
