```python
def change_word(x):
    new_list = []
    for word in x:
        if 'ACS' in word:
            word = 'ACS'
        elif 'American Chemical Society' in word:
            word = 'ACS'
        elif 'AMERICAN CHEMICAL SOCIETY' in word:
            word = 'ACS'
        elif 'ASBMB' in word:
            word = 'ASBMB'
        elif 'Biochemistry and Molecular' in word:
            word = 'ASBMB'
        elif 'BMJ' in word:
            word = 'BMJ'
        elif 'Dartmouth' in word:
            word = 'Dartmouth Journal Services'
        elif 'Darmouth' in word:
            word = 'Dartmouth Journal Services'
        elif 'Elsevier' in word:
            word = 'ELSEVIER'
        elif 'Elseveier' in word:
            word = 'ELSEVIER'
        elif 'Wiley' in word:
            word = 'Wiley'
        elif 'wiley' in word:
            word = 'Wiley'
        elif 'Wliey' in word:
            word = 'Wiley'
        elif 'Oxford University Press'in word:
            word = 'OUP'
        elif 'OXFORD UNIVERSITY PRESS' in word:
            word = 'OUP'
        elif 'Oxford' in word:
            word = 'OUP'
        elif 'Springer' in word:
            word = 'Springer'
        elif 'Wolters' in word:
            word = 'Wolters Kluwer'
        new_list.append(word)
    return new_list     
```
