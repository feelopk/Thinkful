```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
sp_train_total = pd.read_csv('renfe.csv')
```

```python
sp_train_total.head(15)
```

```python
df = []
for i in sp_train_total['start_date']:
    j = i.split(' ')[1]
    df.append(j)

sp_train_total['start_date_time'] = df

df2 = []
for i in sp_train_total['start_date_time']:
    j = i.split(':')[0]
    df2.append(j)
    
sp_train_total['start_date_time_hour'] = df2
```

```python
sp_train_total.groupby('train_type').describe()
sp_train_ave = sp_train_total[sp_train_total['train_type']=='AVE']
sp_train_ave = sp_train_ave[sp_train_ave['price']!='NaN']
sp_train_ave
```

```python
sp_train = sp_train_total.drop(columns=['Unnamed: 0', 'insert_date'])
sp_train.head(15)
```

```python
sp_train_ave.groupby(['origin', 'destination']).describe()
```

```python
sp_train_ave_from_madrid = sp_train_ave[sp_train_ave['origin']=='MADRID']
sp_train_ave_from_madrid = sp_train_ave_from_madrid[['destination', 'start_date_time_hour', 'price']]
sp_train_ave_from_madrid.groupby(['destination', 'start_date_time_hour']).describe()
# sp_train_ave_from_madrid_to_bar = sp_train_ave_from_madrid[sp_train_ave_from_madrid['destination']=='BARCELONA']
# sp_train_ave_from_madrid_to_bar.index = sp_train_ave_from_madrid_to_bar['origin']
# sp_train_ave_from_madrid_to_bar.index
# sp_train_ave_from_madrid_to_bar = sp_train_ave_from_madrid_to_bar[['destination','start_date_time_hour','price']]
# sp_train_ave_from_madrid_to_bar
sns.set(style='whitegrid')
plt.figure(figsize=(20,10))
dataframe = sp_train_ave_from_madrid
g = sns.catplot(x="start_date_time_hour", y="price", hue="destination",
                capsize=0.2, palette="YlGnBu_d", height=6, aspect=1,
                kind="point", data=dataframe)
g.despine(left=True)
```

```python
sp_train_time_group = sp_train.groupby('start_date_time').mean()
sp_train_hour_group = sp_train.groupby('start_date_time_hour')
```

```python
sp_train_time_group
sp_train_hour_group
```

```python
x = sp_train_time_group.index
y = sp_train_time_group['price']
time = sp_train['start_date_time_hour']
price = sp_train['price']
madrid_time = list(set(sp_train_ave_from_madrid['start_date_time_hour']))
madrid_time = sorted(madrid_time)
madrid_price = sp_train_ave_from_madrid['price']
dest = list(set(sp_train_ave_from_madrid['destination']))
```

```python
sns.set(style='whitegrid')
plt.figure(figsize=(8,25))

sns.set_color_codes('pastel')
sns.barplot(x=y, y=x, color='b')
plt.xticks(rotation=90)
sns.despine(right=True)
plt.show()
```
