```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
np.random.seed(1221)
x = np.random.rand(250)
df = pd.DataFrame()
df["x"] = x
```

```python
x_mean = np.mean(x)
```

```python
def x_diff(x):
    x_diff_list = []
    for i in x:
        x_diff = i - x_mean
        x_diff_list.append(x_diff)
    return x_diff_list
```

```python
df["x_diff"] = x_diff(x)
```

```python
df["x_diff_sq"] = df["x_diff"] ** 2
```

```python
plt.figure(figsize=(15,10))

plt.subplot(2, 2, 1)
plt.plot(df["x"], color="purple", marker="x")
plt.ylabel("Values")
plt.title("Random Series")

plt.subplot(2, 2, 2)
plt.stem(df["x_diff"], linefmt='-', basefmt='b-', markerfmt='bo')
plt.ylabel("The Difference Values")
plt.title("The Difference Values of Random Series")

plt.subplot(2, 2, 3)
plt.scatter(x=df["x"], y=df["x_diff"], alpha=0.5, color="red")
plt.title("Values VS Difference")

plt.subplot(2, 2, 4)
plt.hist(df["x"], normed=True, color="blue", alpha=0.5, bins=np.arange(-0.5,1,0.01))
plt.hist(df["x_diff"], normed=True, color="red", alpha=0.5, bins=np.arange(-0.5,1,0.01))
plt.title("Normed Values")
plt.show()
```
